# Q7. Show all courses of graphic design where the price is below 100?

```python
df.head(2)
```

```python
def below_100 (x):
    if x == 'Free':
        return 0
    else:
        return int(x)

df['price'] = df.price.apply(below_100)
df[(df.subject == 'Graphic Design') & (df.price < 100)]
```

```python
df[
    (df.subject=='Graphic Design') & (pd.to_numeric(df.price, errors='coerce')<100)
]
```
